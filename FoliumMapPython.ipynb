```python
# Importar librerias
import folium
import pandas as pd
from google.colab import drive
```

```python
# Montar drive
drive.mount('/content/drive')
```

```python
# Leer archivo
data = pd.read_csv("/content/drive/MyDrive/Cursos/Python 2024/stadium.csv", encoding="cp1252")
data.head()
```

```python
latitude = list(data['LAT'])
longitude = list(data['LON'])
name = list(data["NAME"])
capacity = list(data["capacity"])
website = list(data["website"])
picture = list(data['picture'])
```

```python
# Generar mapa
f = folium.FeatureGroup("my map")
for lt, ln, nm, cp, ws, pic in zip(latitude, longitude, name, capacity, website, picture):
  popup_html = f"<b>Name:</b> {nm}<br/>"
  popup_html += f"<b>Capacity:</b> {str(cp)}<br/>"
  popup_html += f'<b>Wikipedia Link: <a href="{ws}" target="_blank">Click here</a></b></br>'
  popup_html += f'<img src="{pic}" height="142" width="290">'
  popup_iframe = folium.IFrame(width=300, height=220, html=popup_html)
  f.add_child(folium.Marker(location=[lt, ln], popup=folium.Popup(popup_iframe), icon=folium.Icon(color="green")))
```

```python
# Mostrar mapa
map = folium.Map(location=[21.1458,79.0082],zoom_start=5)
map.add_child(f)
map
```

```python
# Leer archivo
data = pd.read_csv("/content/drive/MyDrive/Cursos/Python 2024/uscities.csv", encoding="cp1252")
data.head()
```

```python
latitude = list(data['lat'].head(100))
longitude = list(data['lng'].head(100))
city = list(data["city"].head(100))
population = list(data["population"].head(100))
state = list(data["state_name"].head(100))
```

```python
# Generar mapa
f = folium.FeatureGroup("my map")
for lt, ln, nm, pop, st in zip(latitude, longitude, city, population, state):
  popup_html = f"<b>Name:</b> {nm}<br/>"
  popup_html += f"<b>State:</b> {st}<br/>"
  popup_html += f"<b>Population:</b> {str(pop)}<br/>"
  popup_iframe = folium.IFrame(width=180, height=80, html=popup_html)
  f.add_child(folium.Marker(location=[lt, ln], popup=folium.Popup(popup_iframe), icon=folium.Icon(color="green")))
```

```python
# Mostrar mapa
map2 = folium.Map(location=[25.784,-80.2101],zoom_start=8)
map2.add_child(f)
map2
```
